```python
df = pd.DataFrame([p, q], index=['p', 'q']).T
```

```python
def simple_hypothesis_testing(distributions, n=200):
    sampling_idx = np.random.choice(2)
    return sampling_idx, np.random.choice(D, p=distributions[sampling_idx], size=n)
    
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(20, 20), sharey=True, sharex=True)

df['q'].plot(ax=axes[0], kind='bar', figsize=(20, 6))
df['p'].plot(ax=axes[1], kind='bar', figsize=(20, 6))
plt.tight_layout()
```
