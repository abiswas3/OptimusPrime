```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set('poster')
```

```python
def steves_equations(n):

    k = 10**5    
    Xs = [x for x in range(10**6, 10**7, 10**5)]
    net_rate = [x for x in range(-20, 20, 2)]    
    
    data = []
    for X in Xs:        
        temp = []
        for rate in net_rate:
            alpha = (1 + rate/100)
            temp.append((X-k)*alpha**(n-1) - sum([alpha**j for j in range(1, n-1)])*k)
        data.append(np.log10(temp))

    df = pd.DataFrame(data, index=Xs, columns=net_rate)
    return df
```

```python
df = steves_equations(20)
ax = sns.heatmap(df)
```
